```python
#MULTIPROCESSING WITH EXAMPLE WHERE 3 NUMBER LISTS ARE TAKEN AND DIFFERENT MATH OPERATIONS ARE APPLIED WHOSE VALUES ARE STROED IN THE LISTS. LATER ON THEIR SPEED IS CHECKED WITH AND WITHOUT THE APPLICATION OF MULTIPROCESSING
import multiprocessing as mp
import time
import math
results_a=[] #RESULT LIST-1
results_b=[] #RESULT LIST-2
results_c=[] #RESULT LIST-3
def make_calculation_one(numbers):
    for number in numbers:
        results_a.append(math.sqrt(number ** 3)) #OPERATION FOR LIST 1 BEING APPENDED TO LIST-1
def make_calculation_two(numbers):
    for number in numbers:
        results_b.append(math.sqrt(number ** 4))  #OPERATION FOR LIST 2 BEING APPENDED TO LIST-2
def make_calculation_three(numbers):
    for number in numbers:
        results_c.append(math.sqrt(number ** 5))  #OPERATION FOR LIST 3 BEING APPENDED TO LIST-3
if __name__ == '__main__': #ESSENTIAL TO CATCH ERRORS IN CASE OF MULTIPROCESSING
    number_list= list(range(1000000))
p1= mp.Process(target=make_calculation_one, args=(number_list,)) #Declaring process 1
p2= mp.Process(target=make_calculation_two, args=(number_list,)) #Declaring process 2
p3= mp.Process(target=make_calculation_three, args=(number_list,)) #Declaring process 3
start= time.time()
p1.start() #Starting process-1
p2.start() #Starting process-2
p3.start() #Starting process-3
end=time.time()
print(end-start)

#temp_a=results_a
#temp_b=results_b
#temp_c=results_c   #(IN ORDER TO CHECK IF THE OUTPUT FOR THE CODE WITH AND WITHOUT MULTIPROCESSING IS SAME, UNCOMMENT THESE TEMP STATEMENTS AND THE ONES COMMENTED AT THE LAST)

start=time.time() #starting the same process one by one without mutliprocessing
make_calculation_one(number_list)
make_calculation_two(number_list)
make_calculation_three(number_list)
end=time.time()

print(end-start)
#print(temp_a==results_a)
#print(temp_b==results_b)
#print(temp_c==results_c)
```

```python
#MULITHREADING
import threading
def function1():
  for i in range (10000):
    print("ONE")

def function2():
  for i in range (10000):
    print("TWO")

t1= threading.Thread(target= function1)

t2= threading.Thread(target= function2)

t1.start()
t2.start()
```

```python
#Here, we illustrate the importance of join function. Join function is used when we want the code to wait until the execution of main thread is completed.
#So, in the following code, due to multithreading, even before the main thread is completed, the execution of rest of the code(in our case, print("waiting for thread")) continues.
import threading
def function1():
  for i in range (10000):
    print("ONE")

t1= threading.Thread(target= function1)
t1.start()

print("Waiting for thread")
```

```python
#So, in the above output we see that the print statement is executed parallely. So, in order to put the executuion on hold until the main thead is executed, we use the JOIN function as illustrated below
import threading
def function1():
  for i in range (10000):
    print("ONE")

t1= threading.Thread(target= function1)
t1.start()

t1.join()
print("Waiting for thread")
```
